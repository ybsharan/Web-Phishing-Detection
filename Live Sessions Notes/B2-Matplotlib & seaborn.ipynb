## Draw multiple graphs in one single figure (subplots)

```python
rcParams['figure.figsize']=10,10
```

```python
fig,ax=plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,1].pie(data)
ax[1,0].bar(x,y)
ax[1,1].scatter(x,y)
plt.show()
```
